```python
import networkx as nx
from matplotlib import pyplot as plt

#creating a graph object for a network
G = nx.Graph()

#add nodes
G.add_node('Cassidy')
G.add_nodes_from(['Comfort', 'Yutong'])

G.add_edge('Cassidy', 'Comfort')
G.add_edges_from([('Comfort', 'Yutong'), ('Yutong', 'Cassidy')])

plt.figure(figsize=(7.5, 7.5))
#draw network G
nx.draw_networkx(G)
```

```python
list(G.nodes)
```

```python
#Attribute = undergraduate completed in the U.S. (0 = no; 1 = yes)
us_undergrad = {"Cassidy": 1, "Comfort": 1, "Yutong": 0}
```

```python
for node_id in G.nodes:
    G.nodes[node_id]["study"] = us_undergrad[node_id]
```

```python
node_color = [
    'red' if G.nodes[v]["study"] == 0
    else 'blue' for v in G]
```

```python
nx.draw_networkx(G, node_color = node_color)
```
